```python
u_np, v_np = forward_model.compute_u_v_np()
u, v = forward_model.compute_u_v()
assert np.array_equal(u.numpy(), u_np)
assert np.array_equal(v.numpy(), v_np)


row_indices_np, col_indices_np = forward_model.compute_shifted_indices_np(width, height)
row_indices, col_indices = forward_model.compute_shifted_indices(width, height)
assert np.array_equal(row_indices.numpy(), row_indices_np)
assert np.array_equal(col_indices.numpy(), col_indices_np)

mask_np = forward_model.generate_mask_np(row_indices_np, col_indices_np, width, height)
mask = forward_model.generate_mask(row_indices, col_indices, width, height)
assert np.array_equal(mask.numpy(), mask_np)

row_np, col_np = forward_model.compute_mask_flattened_indices_np(row_indices_np, col_indices_np, mask_np, width, height)
row, col = forward_model.compute_mask_flattened_indices(row_indices, col_indices, mask, width, height)
assert np.array_equal(row.numpy(), row_np)
assert np.array_equal(col.numpy(), col_np)

# computer
r_np = forward_model.computer(gt_dpt.numpy(), globals.Df, use_torch=False)
r = forward_model.computer(gt_dpt, globals.Df, use_torch=True)
assert np.allclose(r.numpy(), r_np)
assert isinstance(r_np, np.ndarray)
assert torch.is_tensor(r)

# copmuteG
G_np, _ = forward_model.computeG(r_np[...,None,None], u_np, v_np, kernel='gaussian', use_torch=False)
G, _ = forward_model.computeG(r.unsqueeze(-1).unsqueeze(-1), u, v, kernel='gaussian', use_torch=True)
assert np.allclose(G.numpy(), G_np)
assert isinstance(G_np, np.ndarray)
assert torch.is_tensor(G)

G_p_np, _ = forward_model.computeG(r_np[...,None,None], u_np, v_np, kernel='pillbox', use_torch=False)
G_p, _ = forward_model.computeG(r.unsqueeze(-1).unsqueeze(-1), u, v, kernel='pillbox', use_torch=True)
assert np.allclose(G_p.numpy(), G_p_np)
assert isinstance(G_p_np, np.ndarray)
assert torch.is_tensor(G_p)
```

```python
pad = max_kernel_size // 2
width, height, _ = aif.shape

gt_aif_tv_map = np.zeros((width, height))
aif_tv_map = np.zeros((width, height))
gt_tv_map = np.zeros((width, height))
tv_map = np.zeros((width, height))

for i in range(pad, width - pad):
    for j in range(pad, height - pad):
        gt_window = gt_aif.cpu()[i - pad:i + pad + 1, j - pad:j + pad + 1]
        gt_aif_tv_map[i, j] = coordinate_descent.total_variation_torch(gt_window).item()
        window = aif.cpu()[i - pad:i + pad + 1, j - pad:j + pad + 1]
        aif_tv_map[i, j] = coordinate_descent.total_variation_torch(window).item()
        
        gt_window = gt_dpt.cpu()[i - pad:i + pad + 1, j - pad:j + pad + 1]
        gt_tv_map[i, j] = coordinate_descent.total_variation_torch(gt_window).item()
        window = dpt.cpu()[i - pad:i + pad + 1, j - pad:j + pad + 1]
        tv_map[i, j] = coordinate_descent.total_variation_torch(window).item()

vmin = min(aif_tv_map.min(), gt_aif_tv_map.min())
vmax = max(aif_tv_map.max(), gt_aif_tv_map.max())
utils.plot_compare_greyscale((aif_tv_map - vmin) / (vmax-vmin), (gt_aif_tv_map - vmin) / (vmax-vmin))
# plt.title('TV of AIF')
plt.show()

vmin = min(tv_map.min(), gt_tv_map.min())
vmax = max(tv_map.max(), gt_tv_map.max())
utils.plot_compare_greyscale((tv_map - vmin) / (vmax-vmin), (gt_tv_map - vmin) / (vmax-vmin))
# plt.title('TV of depth maps')
plt.show()
```
